```python
from src.RAG.indexing import FAISSIndexer
from src.RAG.retrieval import FAISSRetriever
import os
```

## Full usage example of FAISS indexing

```python
Doc_1 = "Water margin is one of the four great classical novels of Chinese literature."
Doc_2 = "Once upon a time, there was a cow."
Doc_3 = "The quick brown fox jumps over the lazy dog."

indexer = FAISSIndexer(dimension=768, use_gpu=True) # Initialize indexer (database) with GPU support

index_path = "data/store_index/faiss_index.index"

if os.path.exists(index_path):
    indexer.load_index(index_path) # Load existing index if available
    # And then you can add more documents, not included retrieval
else:
    indexer.add_documents([Doc_1, Doc_2, Doc_3]) # Add documents to the index
    indexer.save_index(index_path) # Save the index for future use
```

## Full usage example of FAISS retrieving

```python
query = "Tell me a story about an animal."

retriever = FAISSRetriever(index_file=index_path, use_gpu=True) # Initialize retriever with GPU support

results = retriever.retrieve(query, top_k=2)

print("Top retrieved document:", results[0])
```
